## Import the network

## For the ECM: Fibronectin and Laminin (https://amigo.geneontology.org/amigo/gene_product/UniProtKB:P02751, https://amigo.geneontology.org/amigo/gene_product/UniProtKB:P11047)

```python
fibronectin = "P02751"
laminin = "P11047"
```

## Connecting using connect component: component A are my inputs, component B are the nodes of the network in Uniprot format.

```python
new_net1.connect_component(comp_A=[laminin, fibronectin], comp_B=new_net1.nodes['Uniprot'].to_list(), maxlen=1, mode='OUT', only_signed=True, consensus=True)
```

## Fetching genes using the ontology module of NeKo

```python
ontology = Ontology()
```

```python
dna_damage = ontology.get_markers(phenotype="DNA damage sensor activity", id_accession="GO:0140612")
```
